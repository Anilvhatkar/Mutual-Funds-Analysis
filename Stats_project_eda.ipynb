```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('comprehensive_mutual_funds_data.csv')
```

```python
a = sns.countplot(data=dataset, x='min_sip')
for p in a.patches:
    a.annotate(str(p.get_height()), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='bottom')
plt.title('Min Sip')
plt.show()
```

```python
plt.figure(figsize=(10,4))
b = sns.countplot(data=dataset, x='min_lumpsum')
for p in b.patches:
    b.annotate(str(p.get_height()), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='bottom')
plt.title('Min Lumpsum')
plt.show()
```
